```python
urls = []
for link in links:
  urls.extend(re.findall('href="(.*?)"' , str(link)))
len(urls)
```

```python
urls = ['https://www.hindawi.org/'+ s  for s in urls]
len(urls)
```

```python
total = []
wrng_urls = 0
for url in urls:
  try:
    response = requests.get(url, headers=headers)
    # parse the HTML response with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    links = soup.findAll('div' , attrs={'class':'content'}) 
    pattern = r'\b[\u0600-\u06FF]+\b'
    matches = re.findall(pattern, str(links))
    total.extend(matches)
  except:
    print(url)
    wrng_urls = wrng_urls+1

print(wrng_urls)

x = ' '.join(total)

file_name = "books.txt"

with open(file_name, "w") as file:
    file.write(str(x))
```

## Retrieve all content within the same website and save it into a text file

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

# Replace "http://example.com" with the URL of the website you want to scrape
url = "https://www.marefa.org/%D8%A7%D9%84%D8%B5%D9%81%D8%AD%D8%A9_%D8%A7%D9%84%D8%B1%D8%A6%D9%8A%D8%B3%D9%8A%D8%A9"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    try:
        soup = BeautifulSoup(response.content, "html.parser")

        # Get all the links enclosed in <a> tags
        links = soup.find_all("a")

        # Create a variable to store the content
        content = ""

        # Loop over the links and retrieve the content of those within the website
        for link in links:
            # Get the URL of the link
            link_url = link.get("href")

            # Parse the link URL using urlparse
            link_parsed = urlparse(link_url)

            # Check if the link is within the website
            if link_parsed.netloc == urlparse(url).netloc:
                # If the link URL doesn't contain a scheme, append "https://"
                if not link_parsed.scheme:
                    link_url = urljoin(url, link_url)

                # Send a GET request to the link URL
                link_response = requests.get(link_url)

                # Check if the request was successful
                if link_response.status_code == 200:
                    # Parse the HTML content of the link using BeautifulSoup
                    link_soup = BeautifulSoup(link_response.content, "html.parser")

                    # Get all the HTML content enclosed in the <html> tag
                    link_html_content = str(link_soup.html)

                    # Append the HTML content to the variable
                    content += link_html_content
                else:
                    # If the request was unsuccessful, print an error message
                    print(f"Error: Unable to retrieve content of link {link_url}.")
            else:
                # If the link is external, ignore it
                pass

        # Write the content to a file
        with open("output.txt", "a") as f:
            f.write(content)
    except:
        pass
else:
    # If the request was unsuccessful, print an error message
    print("Error: Unable to retrieve website content.")
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import json
```

```python
def getdata(url):
    r = requests.get(url)
    return r.text
```

```python
# create empty dict
dict_href_links = {}

def get_links(website_link):
    html_data = getdata(website_link)
    soup = BeautifulSoup(html_data, "html.parser")
    list_links = []
    for link in soup.find_all("a", href=True):

# Append to list if new link contains original link
        if str(link["href"]).startswith((str(website_link))):
            list_links.append(link["href"])

# Include all href that do not start with website link but with "/"
        if str(link["href"]).startswith("/"):
            if link["href"] not in dict_href_links:
                print(link["href"])
                dict_href_links[link["href"]] = None
                link_with_www = website_link + link["href"][1:]
                print("adjusted link =", link_with_www)
                list_links.append(link_with_www)

# Convert list of links to dictionary and define keys as the links and the values as "Not-checked"
    dict_links = dict.fromkeys(list_links, "Not-checked")
    return dict_links
```

```python
links = get_links("https://www.marefa.org/%D8%A7%D9%84%D8%B5%D9%81%D8%AD%D8%A9_%D8%A7%D9%84%D8%B1%D8%A6%D9%8A%D8%B3%D9%8A%D8%A9")
```

```python

# Create an empty string to hold the combined HTML content
combined_html = ""

# Loop over the URLs in the list
for url in list(links.keys()):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")

        # Get all the HTML content enclosed in the <html> tag
        html_content = str(soup.html)

        # Append the HTML content to the combined HTML string
        combined_html += html_content
    else:
        # If the request was unsuccessful, print an error message
        print("Error: Unable to retrieve website content for " + url)

# Write the combined HTML content to a text file
with open("combined_content.txt", "w") as file:
    file.write(combined_html)

print("HTML content written to file.")
```

```python
# Initialize a counter variable
counter = 1

# Loop over the URLs in the list
for url in list(links.keys()):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")

        # Get the content of the tag that appears in the browser
        tag_content = soup.title.string

        # Create a filename using the tag content and the counter variable
        filename = f"{tag_content}_file{counter}.html"

        try:
            # Write the HTML content to a file with the filename
            with open(filename, "w", encoding="utf-8") as file:
                file.write(str(soup))
            print(f"HTML content written to file: {filename}")

        except Exception as e:
            # If an error occurs while writing the file, print an error message
            print(f"Error writing to file {filename}: {e}")

        # Increment the counter variable
        counter += 1
    else:
        # If the request was unsuccessful, print an error message
        print("Error: Unable to retrieve website content for " + url)
```
